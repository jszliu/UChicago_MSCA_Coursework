```python
## Websraping
import re
import csv
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.m.wikipedia.org/wiki/List_of_current_United_States_senators#List_of_senators'
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko)"}
response = requests.get(url, headers=headers)
html = response.content 
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
senator_table = soup.find('table', attrs={'id': 'senators'})
```

```python
senators = open('us_senators.csv', 'w', encoding='utf-8', newline='')
writer = csv.writer(senators)
```

```python
for idx, tr in enumerate(senator_table.find_all('tr')):
    row = []
    for td in tr:
        value = re.findall('>([^><]+)<', str(td))
        if len(value) > 0:
            row.append(''.join(value).strip())
    if idx > 0:
        if idx % 2 == 0:
            del row[2]
            row = [''] + row
        else:
            del row[3]
    writer.writerow(row)
```

```python
senators.close()
```
