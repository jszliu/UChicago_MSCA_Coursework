# LIWC AUTOMATION

```python
class LIWC:
    def __init__(self,DCT = {"Twitter_id":[] },lock = None,
                init_driver = True ,
                drive_path = 'C:\Program Files (x86)\chromedriver.exe'
                ) -> None:
        self._path = 'C:\Program Files (x86)\chromedriver.exe'
        SER = Service(drive_path)
        chrome_options = Options()
        self.driver = webdriver.Chrome(service=SER, options=chrome_options)
        self._DCT = DCT
        if lock is None:
            lock = threading.Lock()
        self.lock = lock 
        if init_driver:
            self.Init_driver()

    def Init_driver(self):
        website = 'https://liwc.wpengine.com/'
        self.driver.get(website)

    def Analyze_List(self,Text_LST,id_LST,Update = True):
        assert len(Text_LST) == len(id_LST),'Text list must have same length as index'
        for text,id in zip(Text_LST,id_LST):
            self.Analyze(text,id,Update=Update)
            self.driver.back()

    def Analyze(self,text,id,text_cat_idx = 4,Update = True):
        select = Select(self.driver.find_element('xpath','/html/body/div[9]/div/div[2]/div/form/div[1]/select'))
        select.select_by_index(text_cat_idx) #4:Social Media
        Entry_box = self.driver.find_element('xpath','/html/body/div[9]/div/div[2]/div/form/textarea')
        Entry_box.clear()
        Entry_box.send_keys(text)
        #Hit Analyze btn
        Analyze_btn = self.driver.find_element('xpath','/html/body/div[9]/div/div[2]/div/form/center/input')
        Analyze_btn.click()
        #Update Database
        if Update:
            self.Fetch_Res(id)

    def Fetch_Res(self,id):
        #Find Results
        tab = self.driver.find_element('xpath','/html/body/div[9]/table')
        #Parse Result 
        self._DCT['Twitter_id'].append(id)
        for idx,i in enumerate(tab.text.split('\n')):
            self.lock.acquire() #LOCK-----------------<
            if i != 'TRADITIONAL LIWC DIMENSION YOUR DATA AVERAGE FOR SOCIAL MEDIA: TWITTER, FACEBOOK, BLOG' and i!='SUMMARY VARIABLES':
                feature = reduce(str.__add__,(tab.text.split('\n')[idx].split(' ')[:-2]))
                value = tab.text.split('\n')[idx].split(' ')[-2]
                #print(f"{feature}|{value}")

                if feature not in self._DCT:
                    self._DCT[feature] = [value]
                else:
                    self._DCT[feature].append(value)
            self.lock.release() #RELEASE-----------------<

    def Get_DF(self):
        return pd.DataFrame(self._DCT)
```

```python
class LIWC_Manager:
    def __init__(self,n):
        self._DCT = {"Twitter_id":[] }
        self.lock = threading.Lock()
        self.Bots = {idx:LIWC(self._DCT,self.lock,False) for idx in range(n)} 
        #Initialize Threading 
        Threads = [threading.Thread(target=self.Bots[idx].Init_driver, args=()) for idx in range(n)]
        for t in Threads:
            t.start()

    def distribute(self,LST_idx,n):
        from math import ceil
        DCT = {}
        start = 0 
        for i in range(n):
            end = ceil(start+(len(LST_idx)/n))
            DCT[i] = LST_idx[start:end]
            start = end
        return DCT 

    def Analyze(self,df,Col_name='Text'):
        #Divide up the work (index of the df)
        N_bot = len(self.Bots)
        Work_load = self.distribute(df.index,N_bot)
        print(Work_load)
        #Threading 
        threads = [threading.Thread(target=self.Bots[idx].Analyze_List,
             args=(df.iloc[Work_load[idx]][Col_name],Work_load[idx])) 
                for idx in range(N_bot)]
        #Start 
        for t in threads:
            t.start()

    def Get_DF(self):
        return pd.DataFrame(self._DCT)
```

```python
Manager = LIWC_Manager(4)
```

```python
Manager.Analyze(df)
```

```python
res = Manager.Get_DF()
res = res.sort_values('Twitter_id').reset_index()
```
